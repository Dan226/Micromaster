```python

import numpy as np
import matplotlib.pyplot as plt # import the library
import statistics

Xs = np.array([0.0339, 0.0423, 0.213, 0.257, 0.273, 0.273, 0.450, 0.503, 0.503, \
0.637, 0.805, 0.904, 0.904, 0.910, 0.910, 1.02, 1.11, 1.11, 1.41, \
1.72, 2.03, 2.02, 2.02, 2.02])

Ys = np.array([-19.3, 30.4, 38.7, 5.52, -33.1, -77.3, 398.0, 406.0, 436.0, 320.0, 373.0, \
93.9, 210.0, 423.0, 594.0, 829.0, 718.0, 561.0, 608.0, 1.04E3, 1.10E3, \
840.0, 801.0, 519.0])

N = 24

#Media
X = np.mean(Xs)
Y = np.mean(Ys)

print(X,Y)
```

```python
#Desviacion estandar
Xdesv = statistics.stdev(Xs)
Ydesv = statistics.stdev(Ys)

print(Xdesv, Ydesv)
```

```python
#Covarianza
covarianza = np.cov(Xs,Ys)[0,1] #Esto es una matriz 2x2 por lo que con 0,1 se está extrayendo el valor de la covarianza el cual se encuentra en esa posición.
print(covarianza)
```

```python
#Coeficiente de correlación
corrcoef = np.corrcoef(Xs, Ys)[0,1]
print(corrcoef)
```

```python
#Pendiente del modelo linea B1=r*desviaciony/desviacionx

B1 = corrcoef*(Ydesv/Xdesv)
print(B1)
```

```python
#Variable Bo=Ymean - B1*Xmean
B0 = Y - B1*X
print(B0)
```
